'ability_list' is your chosen ability bar, you need to edit this. Abilites are all named fully with no spaces with capital letters at the start of each word ie: GreaterConcentratedBlast as written above. All other forms of punctuation such as apostrophies are removed as well. When addin additonal abilites add a *,* between each one.
This are no checks to ensure your bar is viable, ie 2h vs DW abilities

```python
#ability_list = [Wrack, Impact, GConc, Asphyxiate, Omnipower, DBreath, Auto]
#ability_list = [Sunshine, Asphyxiate, GreaterConcentratedBlast ,DragonBreath, Wrack, Impact, Auto]
ability_list = [GreaterSunshine, GreaterConcentratedBlast, CorruptionBlast, WildMagic , Tsunami,
                DragonBreath,Asphyxiate,Wrack,Combust,Sacrifice,Chain]
```

```python
#@title Main Code Body
ability_list.append(Auto)
 
def select(c_adren):
    adren_check = []
    cd_list = []
    available = []

    
    for ability in ability_list:
        if c_adren >= ability.adren_req:

            adren_check.append(ability)
    
    for ability in adren_check:
        cd_list.append(float(ability.cd))
        available.append(ability)
    print("cd's:", cd_list)
    
    first0 = cd_list.index(0)
    first_ab = available[first0]
    if first_ab.type_ == "Ultimate Buff":
        Player.buff = first_ab.buff
        Player.buff_timer = first_ab.duration
        
    if first_ab.type_ == "Tsunami":
        Player.adren_buff = 10
        Player.adren_buff_timer = 30/0.6
        

    return first_ab


def activate(ability):
    ability.cd = ability.cooldown 
    
    return ability.damage, ability.adren_gain, ability.cast_time

def update():
    for ability in ability_list:
        if ability.cd > 0.:
            ability.cd = round(ability.cd - 1 ,1)
    if Player.buff_timer > 0:
        Player.buff_timer = round(Player.buff_timer - 1 ,1)
        if Player.buff_timer == 0:
            Player.buff = 1.0
    if Player.adren_buff_timer > 0:
        Player.buff_timer = round(Player.buff_timer - 1 ,1)
        if Player.adren_buff_timer == 0:
            Player.adren_buff = 0

def CritCheck():
    if Player.adren_buff_timer > 0:
        crit_chance = Player.critial_hit_chance*100
        num1 = rand.randint(0,100000) #100,000
        if num1 <= crit_chance:
            print("CRITICAL HIT!")
            return 10
        else:
            return 0
    else:
        return 0


def cycle():
    adren =  Player.adrenaline
    total_damage = Player.t_damage
    
    
    ability_used = select(adren) #this selects the first ability on the list not on cooldown.
    print("ability chosen:", ability_used.name)
    
    dmg_done, adren_gain, cast_time = activate(ability_used) #this activates the selected ability and sets in on cooldown
    
    Player.t_damage += dmg_done * Player.buff
    Player.adrenaline += adren_gain + CritCheck()
    if Player.adrenaline > 100:
        Player.adrenaline = 100
    print("adren:", Player.adrenaline, "damage", Player.t_damage)
    
    new_row = {"ability":ability_used.name, "frequency":float(1), "total damage":(dmg_done * Player.buff)}
    return cast_time, new_row

def reset():
    Player.t_damage = 0
    Player.adrenaline = 0
    Player.abiliy_dmg = 2300
    Player.buff = 1.0
    Player.buff_timer = 0
    
tick = 0
cast_cd = 0
df_damage = pd.DataFrame(columns=["ability", "frequency", "total damage"])


for tick in range(final_tick+1):
    if cast_cd == 0:
        print("\ntick:", tick)   #debuggin info
        cast_cd, new_row = cycle()
        df_damage = df_damage.append(new_row, ignore_index=True)
    elif cast_cd > 0:
        cast_cd -= 1
        
    update()
#
#ab_used = pd.DataFrame(abilities_used, columns=["ability"])
#print(ab_used)


#freq_list = ab_used["ability"].value_counts()
#print("\n",freq_list)
def call_damage(table_row):
    print(table_row[0])
    ab_name = table_row.name
    ability_dmg = globals()[ab_name].damage # eval(name).damage
    return ability_dmg


#print(df_damage)
def cleanup(df):
    df = df.groupby("ability").sum()
    df.sort_values(by=["total damage"],inplace=True)
    sum1 = df["total damage"].sum()
    df["% breakdown"] = np.round((df["total damage"]*100)/sum1,2)
    return df
display_table = cleanup(df_damage)
display(df_damage)
```
